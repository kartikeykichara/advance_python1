```python
# the apiriori algorithm association rule learning method that is used to find frequent pattern associations or correlations 
# among set of items in large datasets 
```

```python
# exmaple- in a supermarket if a customer buys a bread and milk they also buys butter 
```

```python
# we find 3 things in this 
#  find support find items and items apperar in the dataset

# support of a = no . of transtactions containg a / total no of transactions
```

```python
# 2. confidence 
# open item b apperas in trasactions that contain item a
# then confi       support of a union b / support of b

```

```python

# 3. Lift
# # when a 

# # lift of a -> b = confidence of a->b / support of b
```

```python
# if lift is >1 then a and b are positively correlated
# if lift <1 then  a and b are negatively  correlated
# if lift==1 then a and b are independent 
```

```python
# Data processing
transactions=[]
for i in range(0,7501):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])

# training the dataset
from apyori import apriori
rules=apriori(transactions,
              min_support= 0.003, 
              min_confidence=0.2,
              min_lift=3,
              min_length=2)

# visualizing the results
a=list(rules)
result=[list(a[i][0]) for i in range(0,len(a))]
```
