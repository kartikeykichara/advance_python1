```python
# #Encoding --->Transform categorial data into numerical data

# 1. Label Encoding---> apply on single column and here we can not decide for
# which sub -category what will be the number.

# 2.Ordinal Encoding----> apply on multiple columns and here decide sub-category value.

# 3.OneHot Encoding ---> it will create new columns based on the sub categories 
# 4.get_dummies------> it will create new columns based on the sub- categories/ 
```
