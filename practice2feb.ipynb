```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\anski\\Desktop\\all_folder\\Datasets\\vgsales.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['Name']=lb.fit_transform(df['Name'])
df['Publisher']=lb.fit_transform(df['Publisher'])
df['Platform']=lb.fit_transform(df['Platform'])
df['Genre']=lb.fit_transform(df['Genre'])
```

```python
df.head()
```

```python
a=df.copy()
a
```

```python
a.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
```

```python
a['Year']=si.fit_transform(a[['Year']])
```

```python
a.isnull().sum()
```

```python
a=a.astype(int)
a
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=a.drop(columns=['Global_Sales'])
y=a['Global_Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
```

```python
x_train_mn=mn.fit_transform(x_train)
```

```python
x_train_mn_new=pd.DataFrame(x_train_mn,columns=x_train.columns)
```

```python
np.round(x_train_mn_new.describe(),2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
